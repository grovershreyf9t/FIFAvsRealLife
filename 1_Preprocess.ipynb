```python
import pandas as pd
import numpy as np
pd.set_option('max_columns',1000)
```

```python
df_fifa_real = pd.read_csv('data_hack.csv')
```

```python
df_fifa_real.head()
```

```python
# Dropping unnecessary columns. Note: To make the model fit closely to objective (to understand how real life stats
# are successful in predicting FIFA ratings), I have considered minimal attribute information from FIFA (only those 
# features that have needed information which is not available in the real world data)
df_fifa_real.drop(['Sr','age','nationality','pac','sho','pas','dri','def','phy','international_reputation','skill_moves','weak_foot',
                   'work_rate_att','work_rate_def','preferred_foot','crossing','finishing','heading_accuracy','volleys','curve','free_kick_accuracy',
                   'ball_control','acceleration','agility','reactions','balance','shot_power','jumping','stamina','strength','long_shots','aggression',
                   'positioning','vision','penalties','composure','marking','standing_tackle','sliding_tackle','gk_handling',
                   'gk_kicking','gk_positioning','gk_reflexes','Points-16/17','Points-15/16','Points-14/15','Points-13/14',
                   'Points-12/13','CS_part-16/17','CS_part-15/16','CS_part-14/15','CS_part-13/14','CS_part-12/13','Sub_p-16/17',
                   'Sub_p-15/16','Sub_p-14/15','Sub_p-13/14','Sub_p-12/13','CO-16/17','CO-15/16','CO-14/15','CO-13/14','CO-12/13',
                   'Conceded_1+-16/17','Conceded_1+-15/16','Conceded_1+-14/15','Conceded_1+-13/14','Conceded_1+-12/13','Start-16/17','Start-15/16',
                   'Start-14/15','Start-13/14','Start-12/13','Sub-16/17','Sub-15/16','Sub-14/15','Sub-13/14','Sub-12/13'],axis=1,inplace = True)
                   
```

```python
df_train = df_fifa_real.drop(df_fifa_real.columns[34:],axis=1)
```

```python
# The filtered data is stored for future reference
df_train.to_pickle('relevant_data.pkl')
```

```python
len(df_train)
```

```python
# The position of some players changes with season. This can be observed by counting unique postions of players
df_group = df_train.groupby(['Players']).count()
duplicate_players = df_group.loc[df_group['Position']>1].index
```

```python
df_train.loc[df_train['Players'].str.contains('Adam Smith')]
```

```python
# Merging of players data. This is done by retaining the latest position of the player and changing its earlier 
# positions
for player in duplicate_players:
    df_temp = df_train.loc[df_train['Players']==player]
    idxs = df_temp.index.tolist()
    max_idx = df_temp['Apps-16/17'].idxmax()
    for idx in idxs:
        if idx != max_idx:
            df_train.drop([idx],axis = 0,inplace = True)
    
```

```python
len(df_train)
```

```python
# Filtering players that have data for two seasons
df_train = df_train.loc[(df_train['Apps-16/17']!=0) & (df_train['Apps-15/16']!=0)]
len(df_train)
```

```python
y = df_train['overall'].reset_index(drop = True)
y.to_csv('targets/targets.csv', index = False)
```

```python
y_gk = df_train.loc[df_train['Position'].str.contains('Goalkeeper')]['overall']
y_def = df_train.loc[df_train['Position'].str.contains('Defender')]['overall']
y_mid = df_train.loc[df_train['Position'].str.contains('Midfielder')]['overall']
y_fwd = df_train.loc[df_train['Position'].str.contains('Forward')]['overall']
df_train.head()
```

```python
y_gk.to_csv('targets/gk_targets.csv', index = False)
y_def.to_csv('targets/def_targets.csv', index = False)
y_mid.to_csv('targets/mid_targets.csv', index = False)
y_fwd.to_csv('targets/fwd_targets.csv', index = False)
```

```python
# I have dropped the league column with an assumption that overall rating would be based on personal attributes,
# irrespective of the league that the player plays in
df_train.reset_index(inplace = True)
df_train.drop(['index','overall'], axis = 1, inplace = True)
x = df_train.drop(['league'], axis = 1)
x.to_csv('training_data/train_data.csv', index = False)
x_pass = x.drop(['sprint_speed','interceptions'], axis = 1)
print(x_pass.columns)
x_pass.to_csv('training_data_pass/train_data_pass.csv', index = False)
```

```python
# Dividing each training dataset into position wise datasets
x_gk = df_train.loc[x['Position'].str.contains('Goalkeeper')]
x_def = df_train.loc[x['Position'].str.contains('Defender')]
x_mid = df_train.loc[x['Position'].str.contains('Midfielder')]
x_fwd = df_train.loc[x['Position'].str.contains('Forward')]
```
